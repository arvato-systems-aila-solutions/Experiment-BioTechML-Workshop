```python
import sys

# Check required libraries and their versions
required_libs = {
    "pandas": "2.2",
    "numpy": "2.0",
    "scikit-learn": "1.7.2",
    "matplotlib": "3.9"
}

print("Checking installed libraries...\n")

for lib_name, min_version in required_libs.items():
    try:
        if lib_name == "scikit-learn":
            import sklearn
            lib = sklearn
            actual_name = "sklearn"
        else:
            lib = __import__(lib_name)
            actual_name = lib_name
        
        installed_version = lib.__version__
        print(f"✓ {lib_name}: {installed_version} (required: >={min_version})")
        
        # Basic version comparison
        installed_parts = [int(x) for x in installed_version.split('.')[:3]]
        required_parts = [int(x) for x in min_version.split('.')[:3]]
        
        if installed_parts < required_parts:
            print(f"  ⚠ Warning: Version {installed_version} is below required {min_version}")
    except ImportError:
        print(f"✗ {lib_name}: NOT INSTALLED (required: >={min_version})")
    except Exception as e:
        print(f"✗ {lib_name}: Error checking - {e}")

print("\nLibrary check complete!")
```
